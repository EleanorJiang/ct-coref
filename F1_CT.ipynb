### Method 1:  Use `text` and `predictor`

```python
override = True
# archive_paths = ["/cluster/work/cotterell/ct/data/centering_exp/coref-spanbert-base-2021.1.10.100"]
archive_paths = ["/cluster/work/cotterell/ct/data/centering_exp/coref-spanbert-base-2021.1.6.60",
                "/cluster/work/cotterell/ct/data/centering_exp/coref-spanbert-base-2021.1.9.20"]
#                     "/cluster/work/cotterell/ct/data/centering_exp/coref-spanbert-base-2021.1.9.100"]
# Dependcy:
import pickle, os, glob
os.chdir('/cluster/work/cotterell/ct/code/centering_exp')
import readOntoNotes
with open('documents.data', 'rb') as filehandle:
    documents = pickle.load(filehandle)
# Predict:   
for archive_path in archive_paths:
    for file in os.listdir(archive_path):
        if not file.endswith(".th") or not (file.startswith("model_state_epoch_") or file.startswith("best")):
            continue
        weights_name = file[:-3]
        weights_file = os.path.join(archive_path, file)
        output_file = os.path.join(
            archive_path, "{}_predicted_dicts.data".format(weights_name))
        if os.path.isfile(output_file) and override == False:
            print(output_file, " exists")
            continue
        predictor = readOntoNotes.load_predictor_from_th(
            archive_path, weights_file, cuda_device=1)
        predicted_dicts = readOntoNotes.get_prediction_from_model(
            documents, predictor)
        print(os.path.join(archive_path, weights_file), len(predicted_dicts))
        with open(output_file, 'wb') as filehandle:
            pickle.dump(predicted_dicts, filehandle)
```

# Plot F1-CT Figure

```python
import os, json
import pandas as pd


def Table2_to_tmp_df(table_2_file, metrics):
    with open(table_2_file, 'r') as f1:
        lines = f1.readlines()
        columns, values = [], []
        for line in lines:
            line = line.strip()
            if line.startswith("index") or line.startswith("Table"):
                continue
            li = list(line.split())
            columns.append(li[1])
            values.append(li[-1])
        assert columns == metrics
        return pd.DataFrame([values], columns=metrics, dtype=float)  # Also can add multiple lines at the same time



def add_F1_and_epoch_and_Experimenets_id_to_tmp_df(archive_path, test_metric_file, name, tmp_df):
    with open(test_metric_file, 'r') as f2:
        # Add coref_f1
        data = json.load(f2)
        coref_f1 = data["coref_f1"] * 100
        tmp_df['coref_F1'] = pd.Series([coref_f1])
        # Add epoch
        if name.startswith("model_state_epoch_"):
            name = name.split("model_state_epoch_")[-1]
        tmp_df['epoch'] = pd.Series([name])
        # Add Experimenets_id
        experimenet_id = archive_path.split("/")[-1]
        tmp_df['experimenet_id'] = pd.Series([experimenet_id])
        
        
def add_everything_of_an_th_to_df(archive_path, name, metrics):
    has_corresponding_table_2 = False
    for Table2_file in os.listdir(archive_path):
        if Table2_file.startswith("{}_".format(name)) and  "Table_2" in Table2_file:
            has_corresponding_table_2 = True
            # Sanity Check
            assert "Sample" in Table2_file
            assert name == Table2_file.split("_Sample_")[0], "{} vs {} is not right in {}".format(Table2_file, name, archive_path)
            search_space_size = Table2_file.split("_Table_2")[0].split("_Sample_")[-1]
            table_2_file = os.path.join(archive_path, "{}_Sample_{}_Table_2".format(name, search_space_size))
            tmp_df = Table2_to_tmp_df(table_2_file, metrics)
            add_F1_and_epoch_and_Experimenets_id_to_tmp_df(archive_path, os.path.join(archive_path, file), name, tmp_df)
            return tmp_df
    if has_corresponding_table_2 == False:
        print("Warning: No Table 2 exists for", archive_path, file)  # Please genererate Table 2 first!
    return None
```

```python
# archive_paths = [ "/cluster/work/cotterell/ct/data/centering_exp/coref-spanbert-base-2021.1.5.100"]
search_space_size = 100
metrics = ['%not_nocb', '%cheapness', '%coherence', '%salience','%kp', 'transition']
df = pd.DataFrame(columns=metrics)
for archive_path in archive_paths:
    if "gold"in archive_path:
        table_2_path = os.path.join(archive_path, "Sample_{}_Table_2".format(search_space_size))
        tmp_df = Table2_to_tmp_df(table_2_path, metrics)
        tmp_df['coref_F1'] = pd.Series([100])
        tmp_df['epoch'] = pd.Series(["-"])
        tmp_df['experimenet_id'] = pd.Series(["gold"]) 
        df = df.append(tmp_df, ignore_index=True)
        continue;
    if "pretrain"in archive_path:
        table_2_path = os.path.join(archive_path, "best_Sample_{}_Table_2".format(search_space_size))
        tmp_df = Table2_to_tmp_df(table_2_path, metrics)
        tmp_df['coref_F1'] = pd.Series([78.87])
        tmp_df['epoch'] = pd.Series(["-"])
        tmp_df['experimenet_id'] = pd.Series(["pretrained_large"]) 
        df = df.append(tmp_df, ignore_index=True)
        continue;
    for file in os.listdir(archive_path):
        if not file.endswith("_test_metric.json"):
            continue;
        name = file.split("_test_metric")[0]
        tmp_df = add_everything_of_an_th_to_df(archive_path, name, metrics)
        df = df.append(tmp_df, ignore_index=True)
        
df = df.append([{'coref_F1':83.1,
                 '%not_nocb':82.70,
                  '%cheapness':78.3,
                  '%coherence': 76.3, 
                 '%salience':82.9,
                 '%kp':83.3, 'transition':79.0,
                'epoch':39, 'experimenet_id':'corefQA'}], ignore_index=True)

frames = []
for metric in metrics:   
    tmpdf= df[['coref_F1',metric, 'epoch', 'experimenet_id']]  
    tmpdf = tmpdf.rename(columns={'coref_F1':'coref_F1', metric: "CT_score", 'epoch':'epoch', 'experimenet_id':'experimenet_id'})
    tmpdf['M_type'] = metric
    frames.append(tmpdf)     
source = pd.concat(frames, ignore_index = True)

print("All data points:")    
pd.set_option('display.max_rows', None)
df
```
